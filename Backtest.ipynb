```python
# Allocating 200 dollars to all 200 stocks, 1 dollar each
final_value = 0
for i in df.columns.get_level_values(0).unique():
#     print(i, float(df[(i, 'Close')].iloc[-1]), float(df[(i, 'Close')].iloc[0]))
    final_value += 1*(float(df[(i, 'Close')].iloc[-1])/float(df[(i, 'Close')].iloc[0]))
    if math.isnan(final_value):
        print("NaN encountered!!")
        break
cagr = ((final_value/200)**(251/1098)-1)*100
print(f"CAGR is {cagr:.2f}%")
```

```python
# Buying 1 share each of every stock in question
initial_value = 0
final_value = 0
for i in df.columns.get_level_values(0):
    initial_value += float(df[(i, 'Close')].iloc[0])
    final_value += float(df[(i, 'Close')].iloc[-1])
    
cagr = ((final_value/initial_value)**(251/1098)-1)*100
print(f"CAGR is {cagr:.2f}%")
```

```python
df_a = df.xs(key='Close', level=1, axis=1)
# df_a = df_a[df_a.columns[0:2]]
df_b = df.xs(key='Open', level=1, axis=1)

new_cols = pd.MultiIndex.from_product([df_a.columns, ['Open', 'Close', 'Ret', 'Inc.', '3', 'Log', 'Alpha', 'Neutralized']])

df_alpha = pd.DataFrame(index=df_a.index, columns=new_cols)

for i in df_a.columns:
    df_alpha[(i, 'Close')] = df_a[i].astype(float)
    df_alpha[(i, 'Open')]  = df_b[i].astype(float)
    df_alpha[(i, 'Ret')]   = df_alpha[(i, 'Close')]/df_alpha[(i, 'Open')]
    df_alpha[(i, 'Inc.')]  = df_alpha[(i, 'Close')] > df_alpha[(i, 'Close')].shift(1)
    df_alpha[(i, 'Log')]   = -np.log(df_alpha[(i, 'Close')] / df_alpha[(i, 'Close')].shift(1))
    
    df_alpha[(i, '3')]     = (df_alpha[(i, 'Inc.')] & df_alpha[(i, 'Inc.')].shift(1) & df_alpha[(i, 'Inc.')].shift(2)) | ~(df_alpha[(i, 'Inc.')] | df_alpha[(i, 'Inc.')].shift(1) | df_alpha[(i, 'Inc.')].shift(2))
    df_alpha.loc[df_alpha.index[0], (i, '3')] = False
    df_alpha.loc[df_alpha.index[1], (i, '3')] = False
    df_alpha.loc[df_alpha.index[2], (i, '3')] = False
    
    df_alpha.loc[df_alpha.index[0], (i, 'Alpha')] = 0
    df_alpha.loc[df_alpha.index[1:], (i, 'Alpha')] = np.where(df_alpha.loc[df_alpha.index[1:], (i, '3')], df_alpha[(i, 'Log')].iloc[1:], np.nan)
    df_alpha[(i, 'Alpha')] = df_alpha[(i, 'Alpha')].ffill()
    
    
avg = df_alpha.loc[:, pd.IndexSlice[:, 'Alpha']].mean(axis=1)

for i in df_a.columns:
    df_alpha[(i, 'Neutralized')] = df_alpha[(i, 'Alpha')] - avg

abs_sum = df_alpha.loc[:, pd.IndexSlice[:, 'Neutralized']].abs().sum(axis=1)
mask = abs_sum > 0

for i in df_a.columns:
    df_alpha.loc[mask, (i, 'Neutralized')] /= abs_sum[mask]

df_alpha['Prof'] = 0
df_alpha['Cap'] = 1

for i in df_a.columns:
    df_alpha[('Prof', '')] += df_alpha[(i, 'Ret')]*(df_alpha[(i, 'Neutralized')].shift(1))
    
df_alpha.loc[df_alpha.index[0], ('Prof', '')] = 0
df_alpha.loc[df_alpha.index[0], ('Cap', '')] = 1

for i in range(1, len(df_a.index)):
    c = df_alpha.iloc[i-1][('Cap', '')]
    p = df_alpha.iloc[i-1][('Prof', '')]
    df_alpha.loc[df_alpha.index[i], ('Cap', '')] = (1+p)*c
```
